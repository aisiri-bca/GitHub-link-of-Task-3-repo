```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
df = pd.read_csv('titanic.csv')

# Select relevant columns and drop missing data
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].dropna()

# Encode categorical features
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Define X and y for modeling
X = df.drop("Survived", axis=1)
y = df["Survived"]

# 🎯 1. Pie Chart of Survival Rate
survived_counts = df['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
colors = ['lightcoral', 'mediumseagreen']
plt.figure(figsize=(6,6))
plt.pie(survived_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('1️⃣ Survival Rate')
plt.show()

# 👨‍👩‍👧‍👦 2. Gender-wise Survival
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.xticks([0,1], ['Male', 'Female'])
plt.title('2️⃣ Survival by Gender')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# 🎂 3. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', kde=True, hue='Survived', multiple='stack', palette='coolwarm')
plt.title('3️⃣ Age Distribution of Passengers')
plt.show()

# 🛳️ 4. Class-wise Survival
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title('4️⃣ Survival by Passenger Class')
plt.xlabel("Pclass (1 = 1st class)")
plt.ylabel("Count")
plt.show()

# 🔥 5. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('5️⃣ Feature Correlation Heatmap')
plt.show()

# 🌲 6. Feature Importance (Random Forest)
model = RandomForestClassifier()
model.fit(X, y)

importances = model.feature_importances_
features = X.columns

plt.figure(figsize=(8,6))
sns.barplot(x=importances, y=features, palette='magma')
plt.title('6️⃣ Feature Importance from Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.show()
```
